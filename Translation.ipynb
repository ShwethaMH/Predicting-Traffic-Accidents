```python
import googletrans
from googletrans import Translator
```

```python
translator = Translator()
```

```python
df_en = df.copy()
```

```python
df_en.rename(columns=lambda x: translator.translate(x).text, inplace=True)
```

```python
df_en.columns
```

```python
df_en.rename(columns = {'A. Simple injuries':'No. simple injuries',
                     'A. He severe injuries':'No. severe injuries',
                     'P. death injuries':'No. death injuries',
                     'A. Medium injuries':'No. medium injuries',
                    'The nationality of the vehicle': "Vehicle's origin",
                     'Omar driver': "Driver's age",
                     'Driver': "Driver's gender",
                    'The name': 'Name',
                    'the speed': "Speed",
                     "The driver's mistake": "Driver's mistake",
                    "The number of vehicles": "No. vehicles",
                     'Vehicle': "Vehicle's type",
                        
                    }, inplace = True)
```

```python
df_en.columns
```

```python
def translate(columns):
    # we create a dictionary where the keys are the terms in English and the values are the original terms in Arabic
    translations = {}
    for column in columns:
        # unique elements of the column
        unique_elements = df_en[column].astype(str).unique()
        for element in unique_elements:
            # add translation to the dictionary
            translations[element] = translator.translate(element).text
    return translations 
```

```python
Columns_to_trens = ['Type','Road lanes', 'Road surface', "Vehicle's origin", 'Class license',
       'Road type', 'Light', 'the weather', 'Road properties',
       "Driver's mistake", 'Conservative', "Vehicle's type", "Driver's gender"]
```

```python
for column in Columns_to_trens:
    print(column)
    print(df_en[column].isnull().sum())
    print("----------------")
```

```python
translations = translate(Columns_to_trens)
```

```python
translations
```

### Before changing the data frame, we check the translations made by googletrans

```python
# modify wrong translation
translations['Mention'] = 'Male'
translations['feminine'] = 'Female'
```

```python
# modify all the terms of the data frame by using the previously created dictionary
df_en.replace(translations, inplace = True)

# check translation
df_en.head()
```
